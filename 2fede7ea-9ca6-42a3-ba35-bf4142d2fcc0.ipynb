````python
def total_sale(row):
    na = row['na_sales']
    eu = row['eu_sales']
    jp = row['jp_sales']
    ot = row['other_sales']
    
    total = na + eu + jp + ot
    return total

df['total_sales'] = df.apply(total_sale, axis=1)
df[df['rating'].isna()]
````
